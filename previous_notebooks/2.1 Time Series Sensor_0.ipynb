```python
#importing all the modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline
```

```python
sensor = pd.read_csv('sensor0.csv')
```

```python
sensor.drop(['Unnamed: 0','dt'],axis=1, inplace=True)
```

```python
sensor.head()
```

```python
ax = sensor.con.plot(figsize=(10,10), color="IndianRed")
plt.xticks(rotation=75, fontsize=20)
plt.yticks(fontsize=20)
plt.ylim(-20, plt.ylim()[1])
plt.legend(loc ='best', fontsize=20)
plt.xlabel("Hours", fontsize=20)
plt.ylabel("Particulate concentrantion", fontsize=20)
plt.title('Time Series of Particulate Concentrantion in Sensor 0', fontsize=20)
```

**Fig.1:** Time Series of Particulate Concentrantion (particle count per 0.01 cubic feet) by hour measured by sensor 0 located at RHI roof (at corner of Hicks and W 9th streets, elevation approx. 8m) - 767 Hicks St, Brooklyn, NY 11231. It can be seen that there is a spike that has to be detected later in the Detenction event.

```python
sensor.spl.plot(figsize=(10,10), color="IndianRed")
plt.xticks(rotation=75, fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=20)
plt.xlabel("Hours", fontsize=20)
plt.ylabel("Noise - Sound preassure level", fontsize=20)
plt.title('Time Series of Noise Sensor 0', fontsize=20)
```

**Fig.2:** Time Series of noise (sound pressure level in dB) by hour measured by sensor 0 located at RHI roof (at corner of Hicks and W 9th streets, elevation approx. 8m) - 767 Hicks St, Brooklyn, NY 11231. It can be seen that time series is uniform.

```python
# Grouping by day.
```

```python
sensor_con = sensor[['date', 'con']].groupby('date').sum()
sensor_spl = sensor[['date', 'spl']].groupby('date').sum()
```

```python
ax = sensor_con.plot(figsize=(10,10), color="IndianRed")
plt.xticks(rotation=75, fontsize=20)
plt.yticks(fontsize=20)
plt.ylim(-20, plt.ylim()[1])
plt.legend(loc ='best', fontsize=20)
plt.xlabel("Date", fontsize=20)
plt.ylabel("Particulate concentrantion", fontsize=20)
plt.title('Time Series of Particulate Concentrantion in Sensor 0', fontsize=20)
```

**Fig.3:** Time Series of Particulate Concentrantion (particle count per 0.01 cubic feet) by day measured by sensor 0 located at RHI roof (at corner of Hicks and W 9th streets, elevation approx. 8m) - 767 Hicks St, Brooklyn, NY 11231.. It can be seen that there is a spike that has to be properly detected in the Detenction event at 27/06.

```python
sensor_spl.plot(figsize=(10,10), color="IndianRed")
plt.xticks(rotation=75, fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=20)
plt.xlabel("Date", fontsize=20)
plt.ylabel("Noise - Sound preassure level", fontsize=20)
plt.title('Time Series of Noise Sensor 0', fontsize=20)
```

**Fig.4:** Time Series of noise (sound pressure level in dB) by days measured by sensor 0 located at RHI roof (at corner of Hicks and W 9th streets, elevation approx. 8m) - 767 Hicks St, Brooklyn, NY 11231.. It can be seen that time series is uniform, but there is a spikes that has to be detected properly in the event detection analysis.
