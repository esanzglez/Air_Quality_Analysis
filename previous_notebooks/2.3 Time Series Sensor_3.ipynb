```python
#importing all the modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline
```

```python
sensor = pd.read_csv('sensor3.csv')
```

```python
sensor.drop(['Unnamed: 0','dt'],axis=1, inplace=True)
```

```python
sensor.head()
```

```python
ax = sensor.con.plot(figsize=(10,10), color="IndianRed")
plt.xticks(rotation=75, fontsize=20)
plt.yticks(fontsize=20)
plt.ylim(-20, plt.ylim()[1])
plt.legend(loc ='best', fontsize=20)
plt.xlabel("Hours", fontsize=20)
plt.ylabel("Particulate concentrantion", fontsize=20)
plt.title('Time Series of Particulate Concentrantion in Sensor 3', fontsize=20)
```

**Fig.13:** Time Series of Particulate Concentrantion (particle count per 0.01 cubic feet) by hour measured by sensor 3 located at RHI entrance (on Hicks street at street level) - 767 Hicks St, Brooklyn, NY 11231. It can be seen a spike that catched the attention. Later the most relevant will be analysed.

```python
sensor.spl.plot(figsize=(10,10), color="IndianRed")
plt.xticks(rotation=75, fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=20)
plt.xlabel("Hours", fontsize=20)
plt.ylabel("Noise - Sound preassure level", fontsize=20)
plt.title('Time Series of Noise Sensor 3', fontsize=20)
```

**Fig.14:** Time Series of noise (sound pressure level in dB) by hour measured by sensor measured by sensor 3 located at RHI entrance (on Hicks street at street level) - 767 Hicks St, Brooklyn, NY 11231. It can be seen as very uniform

```python
# Grouping by day.
```

```python
sensor_con = sensor[['date', 'con']].groupby('date').sum()
sensor_spl = sensor[['date', 'spl']].groupby('date').sum()
```

```python
ax = sensor_con.plot(figsize=(10,10), color="IndianRed")
plt.xticks(rotation=75, fontsize=20)
plt.yticks(fontsize=20)
plt.ylim(-20, plt.ylim()[1])
plt.legend(loc ='best', fontsize=20)
plt.xlabel("Date", fontsize=20)
plt.ylabel("Particulate concentrantion", fontsize=20)
plt.title('Time Series of Particulate Concentrantion in Sensor 3', fontsize=20)
```

**Fig.15:** Time Series of Particulate Concentrantion (particle count per 0.01 cubic feet) by day sensor 3 located at RHI entrance (on Hicks street at street level) - 767 Hicks St, Brooklyn, NY 11231. There is a spike that it will be detected later on detecton analysis.

```python
sensor_spl.plot(figsize=(10,10), color="IndianRed")
plt.xticks(rotation=75, fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=20)
plt.xlabel("Date", fontsize=20)
plt.ylabel("Noise - Sound preassure level", fontsize=20)
plt.title('Time Series of Noise Sensor 3', fontsize=20)
```

**Fig.16:** Time Series of noise (sound pressure level in dB) by day measured bysensor 3 located atRHI entrance (on Hicks street at street level) - 767 Hicks St, Brooklyn, NY 11231. It can be seen that time series is uniform but it is a drop at the end of the collected data.
