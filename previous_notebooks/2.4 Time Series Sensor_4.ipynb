```python
#importing all the modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline
```

```python
sensor = pd.read_csv('sensor4.csv')
```

```python
sensor.drop(['Unnamed: 0','dt'],axis=1, inplace=True)
```

```python
sensor.head()
```

```python
ax = sensor.con.plot(figsize=(10,10), color="IndianRed")
plt.xticks(rotation=75, fontsize=20)
plt.yticks(fontsize=20)
plt.ylim(-20, plt.ylim()[1])
plt.legend(loc ='best', fontsize=20)
plt.xlabel("Hours", fontsize=20)
plt.ylabel("Particulate concentrantion", fontsize=20)
plt.title('Time Series of Particulate Concentrantion in Sensor 4', fontsize=20)
```

**Fig.19:** Time Series of Particulate Concentrantion (particle count per 0.01 cubic feet) by hour measured by sensor 4 located at PioneerWorks roof (elevation approx. 15m) - 159 Pioneer St, Brooklyn, NY 11231. It can be seen some spikes at the end of the data.

```python
sensor.spl.plot(figsize=(10,10), color="IndianRed")
plt.xticks(rotation=75, fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=20)
plt.xlabel("Hours", fontsize=20)
plt.ylabel("Noise - Sound preassure level", fontsize=20)
plt.title('Time Series of Noise Sensor 4', fontsize=20)
```

**Fig. 20:** Time Series of noise (sound pressure level in dB) by hour measured by sensor measured by sensor 4 located at PioneerWorks roof (elevation approx. 15m) - 159 Pioneer St, Brooklyn, NY 11231. It can be seen that as well it is uniform distributed.

```python
# Grouping by day.
```

```python
sensor_con = sensor[['date', 'con']].groupby('date').sum()
sensor_spl = sensor[['date', 'spl']].groupby('date').sum()
```

```python
ax = sensor_con.plot(figsize=(10,10), color="IndianRed")
plt.xticks(rotation=75, fontsize=20)
plt.yticks(fontsize=20)
plt.ylim(-20, plt.ylim()[1])
plt.legend(loc ='best', fontsize=20)
plt.xlabel("Date", fontsize=20)
plt.ylabel("Particulate concentrantion", fontsize=20)
plt.title('Time Series of Particulate Concentrantion in Sensor 4', fontsize=20)
```

**Fig.21:** Time Series of Particulate Concentrantion (particle count per 0.01 cubic feet) by day sensor measured by sensor 4 located at PioneerWorks roof (elevation approx. 15m) - 159 Pioneer St, Brooklyn, NY 11231.There is a spike to be detected.

```python
sensor_spl.plot(figsize=(10,10), color="IndianRed")
plt.xticks(rotation=75, fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=20)
plt.xlabel("Date", fontsize=20)
plt.ylabel("Noise - Sound preassure level", fontsize=20)
plt.title('Time Series of Noise Sensor 4', fontsize=20)
```

**Fig.22.** Time Series of noise (sound pressure level in dB) by day measured bysensor measured by sensor 4 located at PioneerWorks roof (elevation approx. 15m) - 159 Pioneer St, Brooklyn, NY 11231.It can be seen that there is a visible spike and it drops at the end of the data.
